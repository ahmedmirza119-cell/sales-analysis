```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("Sales Data.csv" ,encoding= 'unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
# drop un related columns
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
# check for null values
pd.isnull(df).sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
# initializing list of lists
data_test = [['madhav',11], ['Gopi',51], ['Keshav',], ['Lalita', 16]]

# create  the pandas dataframe
df_test = pd.DataFrame(data_test, columns=['Name','Age'])
df_test
```

```python
# drop null values
df_test.dropna(inplace=True)
```

```python
df_test
```

```python
# change data types
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
# decribe() method returns description of the data in the dataframe (i.e. count,mean,std,etc)
df.describe()
```

```python
# use desribe for specific column
df[['Age','Orders','Amount']].describe().round()
```

```python
# 'df' aapka cleaned dataframe hai
# index=False zaroori hai taake serial numbers (0,1,2...) alag se column na ban jayein
df.to_csv('Sales Data(cleaned).csv', index=False)

print("File successfully save ho gayi hai: cleaned_data.csv")
```

```python
# Exploratory Data Analysis
```

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
# Corrected version
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Plotting
sns.barplot(x='Gender', y='Amount', data=sales_gen)
```

```python
df.columns
```

```python
ax =sns.countplot(data = df, x='Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)
```

```python
# total numbers of orders in top 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Fixed syntax: added '=', starting quote, and fixed spelling
sns.set(rc={'figure.figsize':(15, 5)})

sns.barplot(data = sales_state, x ='State', y ='Orders')
```

```python
# Total amount sales from Top 10 states
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Fixed syntax: added '=', starting quote, and fixed spelling
sns.set(rc={'figure.figsize':(15, 5)})

sns.barplot(data = sales_state, x ='State', y ='Amount')
```

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(4,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_data = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6, 5)}) # Slightly wider for better readability
# Changed data from sales_state to sales_data
sns.barplot(data = sales_data, x ='Marital_Status', y ='Amount', hue = 'Gender')
```

```python
# occupation
sns.set(rc={'figure.figsize':(20, 5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_data = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20, 5)}) # Slightly wider for better readability
# Changed data from sales_state to sales_data
sns.barplot(data = sales_data, x ='Occupation', y ='Amount')
```

```python
# Product category
sns.set(rc={'figure.figsize':(30, 5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_data = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)

sns.set(rc={'figure.figsize':(10, 5)}) # Slightly wider for better readability
# Changed data from sales_state to sales_data
sns.barplot(data = sales_data, x ='Product_Category', y ='Amount')
```

```python
sales_data = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20, 5)}) # Slightly wider for better readability
# Changed data from sales_state to sales_data
sns.barplot(data = sales_data, x ='Product_ID', y ='Orders')
```
